```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plot
group_names = ['Accuracy', 'Precision', 'Recall']
metric_names = ['RefactoringMiner', 'GumTree Greedy', 'GumTree Simple']
metric_values1 = [1.0, 1.0, 1.0]
metric_values2 = [0.91742736, 0.98474777, 0.92895323]
metric_values3 = [0.91804206, 0.9843147, 0.92998755]

# Set the width of each bar and the spacing between groups of bars
bar_width = 0.2
spacing = 0.1

# Generate the x positions of the bars for each group and metric
x_pos = np.arange(len(group_names))
x_pos_metric1 = x_pos - bar_width - spacing
x_pos_metric2 = x_pos
x_pos_metric3 = x_pos + bar_width + spacing

# Generate the bar plot
fig, ax = plt.subplots()
ax.bar(x_pos_metric1, metric_values1, bar_width, label=metric_names[0])
ax.bar(x_pos_metric2, metric_values2, bar_width, label=metric_names[1])
ax.bar(x_pos_metric3, metric_values3, bar_width, label=metric_names[2])

for i, v1, v2, v3 in zip(x_pos, metric_values1, metric_values2, metric_values3):
    ax.text(i - bar_width - spacing/2, v1 - 0.1, str(v1), ha='center', va='bottom', rotation=90)
    ax.text(i, v2 - 0.1, str(v2), ha='center', va='bottom', rotation=90)
    ax.text(i + bar_width + spacing/2, v3 - 0.1, str(v3), ha='center', va='bottom', rotation=90)

# Add labels and a title to the plot
# ax.set_xlabel('Metrics')
ax.set_ylabel('Values')
ax.set_title('MatchedProgramElements')

# Add x-axis tick labels
ax.set_xticks(x_pos)
ax.set_xticklabels(group_names)
ax.set_ylim([0.6, 1.01])

# Add a legend to the plot
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3)

# Show the plot
plt.show()
```
